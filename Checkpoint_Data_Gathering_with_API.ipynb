```python
# Import the requests package and store your API KEY in variable.
import requests
api_key = "API_KEY"

# Ask the API endpoint of APOD for the astronomy picture of the day. Get then display the image on your notebook.
from IPython.display import Image, display

base_url = "https://api.nasa.gov/planetary/apod"
params = {
    "api_key": api_key
}
response = requests.get(base_url, params=params)
if response.status_code == 200:
    data = response.json()
    image_url = data["url"]

    print("Astronomy Picture of the Day:")
    display(Image(url=image_url))
else:
    print("Error:", response.status_code)
```

```python
# Select 'Asteroids - NeoWs' option. Store the results in a pandas dataframe.
import pandas as pd
import requests
import json

api_key = "API_KEY"
base_url = "https://api.nasa.gov/neo/rest/v1/feed"
params = {
    "api_key": api_key,
    "start_date": "2025-12-01",
    "end_date": "2025-12-07",
}
response = requests.get(base_url, params=params)

asteroids_list = []

if response.status_code == 200:
  data = response.json()["near_earth_objects"]
  # Loop through each date
  for date in data:
    for asteroid in data[date]:
      asteroids_list.append({
          "Asteroid ID": asteroid["id"],
          "Asteroid Name": asteroid["name"],
          "Min Estimated Diameter (km)": asteroid["estimated_diameter"]["kilometers"]["estimated_diameter_min"],
          "Absolute Magnitude": asteroid["absolute_magnitude_h"],
          "Relative Velocity (km/s)": asteroid["close_approach_data"][0]["relative_velocity"]["kilometers_per_second"]
      })
else:
  print("Error:", response.status_code)

# Creating the DataFrame
asteroids_df = pd.DataFrame(asteroids_list)

# Displaying the first few rows of the DataFrame
asteroids_df.head()

# Export to CSV
asteroids_df.to_csv("nasa_asteroids.csv", index=False)
```
